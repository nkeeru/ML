```python
import numpy as np
from sklearn.model_selection import train_test_split

# Advertising Spend in Rupees
X = np.array([[10000], [15000], [20000], [25000], [30000], [35000], [40000]])  # Direct representation in ₹
y = np.array([200, 300, 450, 500, 600, 700, 800])  # Sales (Units Sold)

# Splitting Data (80% Training, 20% Testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training Data:", X_train)
print("Testing Data:", X_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: House Size vs. Price
X = np.array([600, 800, 1000, 1200, 1500]).reshape(-1, 1)  # House sizes
y = np.array([30, 40, 50, 60, 75])  # Prices in Lakhs

# Creating and training the model
model = LinearRegression()
model.fit(X, y) #The fit() method trains the model on your data.

# Predictions for the given house sizes
y_pred = model.predict(X)

# Visualizing the data points and regression line
plt.scatter(X, y, color='blue', label='Actual Data')  # Scatter plot of actual data
plt.plot(X, y_pred, color='red', linestyle='--', label='Regression Line')  # Regression line
plt.xlabel('House Size (sq. ft)')
plt.ylabel('Price (₹ Lakhs)')
plt.title('Linear Regression - House Price Prediction')
plt.show()
```
